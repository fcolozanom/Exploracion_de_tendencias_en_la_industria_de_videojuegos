```python
# Importa los diccionarios
import pandas as pd
```

```python
# Crea un DataFrame
df_games = pd.read_csv('files/datasets/input/games.csv')
print(df_games.info())
```

```python
# Pone en min√∫sculas los nombres de las columnas
df_games.columns = df_games.columns.str.lower()
display(df_games.head())
```

```python
# Convierte la columna year_of_release a tipo Int64 para manejar NaN
df_games['year_of_release'] = df_games['year_of_release'].astype('Int64')  


# Convierte la columna user_score a float y maneja errores
df_games['user_score'] = pd.to_numeric(df_games['user_score'], errors='coerce')

display(df_games.head())
```

```python
# Las columnas en las que los tipos de datos han sido cambiados en el DataFrame df_games son:

#year_of_release: El tipo de dato original era float. Se cambi√≥ a Int64 (entero con posibilidad de valores nulos) utilizando el m√©todo astype('Int64'). Esto se hizo para permitir representar a√±os como enteros y para poder manejar adecuadamente los valores nulos en caso de que no haya un a√±o de lanzamiento registrado para un juego.

#user_score: El tipo de dato original era object (cadena) debido a la presencia de valores como 'tbd'. Se cambi√≥ a float utilizando la funci√≥n pd.to_numeric con errors='coerce'. Esto se hizo para convertir las calificaciones de usuarios en valores num√©ricos de punto flotante, tratando los valores no num√©ricos (como 'tbd') como NaN para permitir operaciones matem√°ticas y an√°lisis de datos
```
